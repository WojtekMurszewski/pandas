```python
import pandas as pd
airlines_data = pd.read_csv('krk-airlines.csv')
airlines = pd.DataFrame(airlines_data)

flights_data = pd.read_csv('krk-flights.csv')
flights = pd.DataFrame(flights_data)

passengers_data = pd.read_csv('krk-passengers.csv')
passengers = pd.DataFrame(passengers_data)
```

List including flight number and destination (two columns)

```python
flights[['flight', 'to']]
```

Flight list with the full name of the airline and the name of the aircraft

```python
flights_and_airlines = pd.merge(flights, airlines, on='airlineid')
flights_and_airlines
```

List of passengers on a flight to London sorted by surname

```python
passengers_and_flights = pd.merge(passengers, flights, on='flight')
passengers_and_flights.query("to == 'London'")[['surname', 'name']].sort_values('surname')
```

List of women flying to Paris ordered by surname

```python
passengers_and_flights.query("to == 'Paris' and gender == 'Female'")[['surname', 'name', 'gender']].sort_values('surname')
```

Number of passengers on the flight to Berlin

```python
passengers_and_flights.query("to == 'Berlin'")['to'].value_counts()
```

Number of men flying from Krakow

```python
passengers_and_flights.query("'from' == 'Krakow' and gender == 'Male'")[['from','gender']].value_counts()
```

Number of passengers for each flight

```python
passengers.groupby('flight')[['flight']].count()
```
