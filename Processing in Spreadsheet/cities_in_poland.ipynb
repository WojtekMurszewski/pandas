```python
import pandas as pd
import matplotlib.pyplot as plt 

cities_data = pd.read_csv("cities-in-poland.csv", sep=';')
cities = pd.DataFrame(cities_data)

cities_data
```

Create chart showing the total number of inhabitants in each province

```python
cities_data[['Province', 'Population']].groupby('Province').sum().plot(kind = 'bar')
```

Create chart showing the area of the largest city in each province

```python
cities_data[['Province', 'City', 'Areakm2']].groupby('Province').max('Areakm2').plot(kind = 'bar')
```
