```python
import pandas as pd 
import matplotlib.pyplot as plt

employees_data = pd.read_csv('employees.csv')
employees = pd.DataFrame(employees_data)
employees[['sex', 'age']].groupby('sex').mean().round(2).plot(kind='bar', title='Average age of women and men')
```
