```python
import pandas as pd
import matplotlib.pyplot as plt

eu_countries_data = pd.read_csv('european-countries.csv')
eu_countries = pd.DataFrame(eu_countries_data)
eu_countries
```

Create a bar chart with the population of Central European countries (sort the data by population)

```python
eu_countries.index = eu_countries['Name']
eu_countries[['Region', 'Population']].query("Region == 'Central Europe'").sort_values('Population').plot(kind='bar')
eu_countries[['Region', 'Population']].query("Region == 'Central Europe'")
```

Create a pie chart presenting the percentage of the regions' area.

```python
eu_countries[['Region','Area']].groupby('Region').sum().plot(subplots=True, kind='pie', autopct = '%1.0f%%', legend=False, ylabel='')
```
